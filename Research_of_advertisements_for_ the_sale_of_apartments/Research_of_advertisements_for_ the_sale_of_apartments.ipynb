## Предобработка данных

### Обработка пропусков

```python
round((data.isna().sum() / data.shape[0]*100), 2).sort_values(ascending=False) # процент пропущенных значений
```

Во многих колонках есть пропуски. Их необходимо обработать.

* `ceiling_height` — высота потолков (м). Почти 40% пропущено, это довольно много и может повлиять  исследование. Лучше заполнить медианным значением.
* `floors_total` — всего этажей в доме. Меньше сотни пропущенных значений, не отразится на исследовании. Можно смело удалить.
* `living_area` — жилая площадь в квадратных метрах (м²). 8% пропусков. Как правило жилая площадь напрямую влияет на цену квартиры. Цена может сильно различаться.  Ее лучше оставить пустым.
* `is_apartment` — апартаменты (булев тип). 88% пропусков. Можно заменить на False. Не повлияет на исследование.
* `kitchen_area` — площадь кухни в квадратных метрах (м²). Почти 10% пропусков. Кухня может влиять на стоимость, лучшие оставить пустым.
* `balcony` — число балконов. Половина пропусков. Если человек не указал число балконов — скорее всего, их нет. Такие пропуски правильно заменить на 0.
* `locality_name` — название населённого пункта. Всего 49 пропусков. Проще всего удалить.
* `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м). Среднее значение здесь не годится. Также оставляем пустым.
* `cityCenters_nearest` — расстояние до центра города (м). Расстояние может быть любым. Также оставляем пустым.
* `parks_around3000` — число парков в радиусе 3 км. Возможно просто не указали, но скорее всего, их нет. Можно заменить на 0.
* `parks_nearest` — расстояние до ближайшего парка (м). Больше половины пропущено. Расстояние может быть любым. Также оставляем пустым.
* `ponds_around3000` — число водоёмов в радиусе 3 км. Возможно просто не указали, но скорее всего, их нет. Можно заменить на 0.
* `ponds_nearest` — расстояние до ближайшего водоёма (м). Больше половины пропущено. Расстояние может быть любым. Также оставляем пустым.
* `days_exposition` — сколько дней было размещено объявление (от публикации до снятия). Его могли снять в тот же день, либо оно до сих пор висит в продаже. Логично оставить пустым.

```python
data = data.dropna(subset=['floors_total','locality_name']).reset_index(drop=True) # удаляю пропуски со сбросом индексов
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median()) # заполняю пропуски медианой
data['is_apartment'] = data['is_apartment'].fillna(False) # заполняю пропуски False

columns_to_replace = ['balcony','parks_around3000','ponds_around3000']
for column in columns_to_replace: 
    data[column] = data[column].fillna(0) # перебор названий столбцов в цикле и замена пропущенных значений на 0

data = data.rename(columns={'parks_around3000':'parks_around_3000',
                          'ponds_around3000':'ponds_around_3000',
                          'cityCenters_nearest':'city_centers_nearest'}) # переименование столбцов

round((data.isna().sum() / data.shape[0]*100), 2).sort_values(ascending=False) # проверка пропущенных значений в %
```

 **Вывод**

Часть пропусков заменена. Как повлияют оставшиеся пропуски на исследование мне еще предстоит выяснить. 

3 столбцов приведены к "змеиному регистру".

В аннотации к исследованию сказано: "Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма." Это значит что часть пропусков было вызвано человеческим фактором, например, не указано количество балконов, площадь жилой площади и кухни. А часть техническим. В частности это касается расстояний. Возможно слишком большие расстояния система не учла, поэтому значения были пропущены. Либо это деревня или небольшой поселок, у которых нет центра. А это значит они вряд ли повлияют на конечное исследование.

## Расчёты и добавление результатов в таблицу

- цена квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры; варианты — первый, последний, другой;
- соотношение жилой и общей площади, а также отношение площади кухни к общей.

Добавляю столбцы:

```python
data['price_square'] = data['last_price']/data['total_area'] # цена квадратного метра
data['weekday'] = pd.to_datetime(data['first_day_exposition'], format='%Y%m%dT').dt.weekday # день недели
data['month'] = pd.to_datetime(data['first_day_exposition'], format='%Y%m%dT').dt.month # месяц
data['year'] = pd.to_datetime(data['first_day_exposition'], format='%Y%m%dT').dt.year # год
```

```python
def floor_group(row): # функция группировки по этажам
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'

data['floor_group'] = data.apply(floor_group, axis=1) # добавление столбца группировки по этажам
```

```python
data['living_area_ratio'] = data['living_area']/data['total_area'] # соотношение жилой и общей площади
data['kitchen_area_ratio'] = data['kitchen_area']/data['total_area'] # отношение площади кухни к общей 
```

 **Вывод**

Добавил новые столбцы, чтобы удобней было анализировать зависимость цены от разных факторов.
